```python
from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer
```

## Quantile transformer

```python
k=1
fig,ax=plt.subplots(nrows=7,ncols=7, figsize=(30,23))
ax=ax.flatten()
quantile={}
xsQuant=np.copy(xs1)
for i in tqdm(range(len(xs1[0]))):
    if scols[i] in scale_cols:
        print(scols[i], i)
        t=QuantileTransformer().fit(xs1[:int(k*len(xs1)),i].reshape(-1,1))
        xsQuant[:int(k*len(xs1)),i]=np.hstack(t.transform(xs1[:int(k*len(xs1)),i].reshape(-1,1)))
        quantile[scols[i]]=t
    ax[i].hist(xsQuant[:int(k*len(xs1)),i], bins=50, density=1, histtype='step');
    ax[i].set(title=scols[i])
# plt.hist(a[:,n], bins=100, density=1);
fig.tight_layout()
```

## Standard scaler

```python
k=1
fig,ax=plt.subplots(nrows=7,ncols=7, figsize=(30,23))
ax=ax.flatten()
standard={}
xs2=np.copy(xs1)
xsStandard=np.copy(xs1)
for i in tqdm(range(len(xs1[0]))):
    if scols[i] in scale_cols:
        print(scols[i], i)
        t=StandardScaler().fit(xs1[:int(k*len(xs1)),i].reshape(-1,1))
        xsStandard[:int(k*len(xs1)),i]=np.hstack(t.transform(xs1[:int(k*len(xs1)),i].reshape(-1,1)))
        standard[scols[i]]=t
    ax[i].hist(xsStandard[:int(k*len(xs1)),i], bins=50, density=1, histtype='step');
    ax[i].set(title=scols[i])
# plt.hist(a[:,n], bins=100, density=1);
fig.tight_layout()
```

## Power transformer (taken after standard scaler)

```python
k=1
fig,ax=plt.subplots(nrows=7,ncols=7, figsize=(30,23))
ax=ax.flatten()
power={}
xs2=np.copy(xsStandard)
for i in tqdm(range(len(xs1[0]))):
    if scols[i] in scale_cols:
        print(scols[i], i)
        t=PowerTransformer().fit(xs2[:int(k*len(xs1)),i].reshape(-1,1))
        xsPower[:int(k*len(xs2)),i]=np.hstack(t.transform(xs2[:int(k*len(xs2)),i].reshape(-1,1)))
        power[scols[i]]=t
    ax[i].hist(xsPower[:int(k*len(xs1)),i], bins=50, density=1, histtype='step');
    ax[i].set(title=scols[i])
# plt.hist(a[:,n], bins=100, density=1);
fig.tight_layout()
```
